```python
# plot the histogram of the three datasets
values0 = mean_sum_0.values


values1 = mean_sum_1.values

values2 = mean_sum_2.values
# color blue
plt.hist(values0, bins=100, alpha=0.5, color='blue')

plt.hist(values1, bins=100, alpha=0.5, color='red')
plt.hist(values2, bins=100, alpha=0.5, color='green')
plt.show()
```

```python
# compute roc curve
from sklearn.metrics import roc_curve, auc
import numpy as np
import matplotlib.pyplot as plt

# get the true labels
y_true = np.concatenate([np.ones_like(values1), np.zeros_like(values0)])
# get the scores
y_score = np.concatenate([values1, values0])
# compute the roc curve
fpr, tpr, thresholds = roc_curve(y_true, y_score)
# compute the area under the curve
roc_auc = auc(fpr, tpr)
# plot the roc curve
plt.plot(fpr, tpr, lw=1, label='ROC (area = %0.2f)' % (roc_auc))
# plot the random classifier
plt.plot([0, 1], [0, 1], '--', color=(0.6, 0.6, 0.6), label='Random')
# set the limits of the axes

```
